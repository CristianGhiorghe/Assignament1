```python
# Define Target parameter (i.e. independent vairable)
y = DD_df.loc[:,['Bed_Type']].values    
```

```python
# Scale the parameter values (PCA need each dimension to have similar magnitudes)
x = StandardScaler().fit_transform(x)
```

```python
# Check the parameters are correctly defined
pd.DataFrame(data = x, columns = features).head()
```

```python
# PCA Projection to 2D
# Create PCA by projecting the 30D parameters onto a 2D circular
pca = PCA(n_components=2)   
```

```python
# rescale X based on the 30D
principalComponents = pca.fit_transform(x)  
```

```python
# Extract Principle Components
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf.head(5)
```

```python
# check bed type 
DD_df[['Bed_Type']].head()
```

```python
# define final DF to be ploted based on bed type
finalDf = pd.concat([principalDf, DD_df[['Bed_Type']]], axis = 1)
# check df head
finalDf.head(5)
```

```python
# plot the figure based on the two bed types
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)


targets = ['Single', 'Double']
colors = ['r', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Bed_Type'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
# print components detals and cumulative explained_variance_ratio_
print ( "Components = ", pca.n_components_ , 
       ";\nTotal explained variance = ", (pca.explained_variance_ratio_.sum() * 100).round(2).astype(str) + ' %'
      )
```
